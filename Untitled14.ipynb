```python
class Classification(nn.Module):
    def traning_function(self, batch):
        images, labels = batch 
        output = self(images)                  # Pretiction output
        loss = F.cross_entropy(output, labels) # calculates Loss of traning data
        return loss
    
    
    def validate_function(self, batch):
        images, labels = batch 
        output = self(images)                    # Generate predictions
        loss = F.cross_entropy(output, labels)   # Calculate loss
        acc = accuracy(output, labels)           # Calculate accuracy
        return {'Loss_validation': loss.detach(), 'Accuracy_validation': acc}
        
    def epoch_function(self, outputs):
        batch_losses = [x['Loss_validation'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()   # Combine losses
        batch_accs = [x['Accuracy_validation'] for x in outputs]
        epoch_acc = torch.stack(batch_accs).mean()      # Combine accuracies
        return {'Loss_validation': epoch_loss.item(), 'Accuracy_validation': epoch_acc.item()}
    
    def end_epoch_function(self, ep, Total_results):
        print("TotalEpoch [{}], traning_loss: {:.4f}, Loss_validation: {:.4f}, Accuracy_validation: {:.4f}".format(
            ep, Total_results['traning_loss'], Total_results['Loss_validation'], Total_results['Accuracy_validation']))
        
def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))
```

```python


## putting the nn.sequential to activate the layers 
```

```python
class SVHN(Classification):
    def __init__(self):
        super().__init__()
        self.cnnNetworkLayers = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 64 x 16 x 16

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 128 x 8 x 8

            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 256 x 4 x 4

            nn.Flatten(), 
            nn.Linear(256*4*4, 1024),
            nn.ReLU(),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Linear(512, 10))
        
    def forward(self, xb):
        return self.cnnNetworkLayers(xb)
```

```python
def get_default_device():
    """Pick GPU if available, else CPU"""
    if torch.cuda.is_available():
        return torch.device('cuda')
    else:
        return torch.device('cpu')
    
def to_devices(data, device):
    """Move tensor(s) to chosen device"""
    if isinstance(data, (list,tuple)):
        return [to_devices(x, device) for x in data]
    return data.to(device, non_blocking=True)

class DeviceDataLoader():
    """Wrap a dataloader to move data to a device"""
    def __init__(self, dl, device):
        self.dl = dl
        self.device = device
        
    def __iter__(self):
        """Yield a batch of data after moving it to device"""
        for b in self.dl: 
            yield to_devices(b, self.device)

    def __len__(self):
        """Number of batches"""
        return len(self.dl)
```

Traning the Model for predicting Street view House Number

```python
@torch.no_grad()
def DataEvaluation_function(graphical_model, validation_dataLoader1):
    graphical_model.eval()
    outputs = [graphical_model.validate_function(batch) for batch in validation_dataLoader1]
    return graphical_model.epoch_function(outputs)

def DataFit_function(number_of_epochs, r, graphical_model, train_loader1,validation_dataLoader1, optional_function=torch.optim.SGD):
    history = []
    optimization = optional_function(graphical_model.parameters(), r)
    for ep in range(number_of_epochs):
        # Training Phase 
        graphical_model.train()
        Tranning_lost_data = []
        for batch in train_loader1:
            batch_loss = graphical_model.traning_function(batch)
            Tranning_lost_data.append(batch_loss)
            batch_loss.backward()
            optimization.step()
            optimization.zero_grad()
        # Validation phase
        Total_results = DataEvaluation_function(graphical_model, validation_dataLoader1)
        print('1',Total_results)
        Total_results['traning_loss'] = torch.stack(Tranning_lost_data).mean().item()
        graphical_model.end_epoch_function(ep, Total_results)
        history.append(Total_results)
    return history
```

```python
Epoch_history = DataFit_function(number_of_epochs, r, graphical_model, train_loader1, validation_dataLoader1, optional_function)
```

```python
def plot_accuracies(Epoch_history):
    accuracies = [x['Accuracy_validation'] for x in Epoch_history]
    plt.plot(accuracies, '-x')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. No. of epochs');
```

```python
plot_accuracies(Epoch_history)
```

```python
def Epoch_vs_Loss(Epoch_history):
    train_losses = [x.get('traning_loss') for x in Epoch_history]
    val_losses = [x['Loss_validation'] for x in Epoch_history]
    plt.plot(train_losses, '-gx')
    plt.plot(val_losses, '-yx')
    plt.xlabel('Total Number of Epochs')
    plt.ylabel('Loss Values')
    plt.legend(['Traning Epoch', 'Validation Epoch'])
    plt.title('Total Loss vs. Individual Total epochs');

   
```

```python
def Individual_image_pridiction(img, graphical_model):
    # Convert to a batch of 1
    xb = to_devices(img.unsqueeze(0), Device)
    # Get predictions from graphical_model
    yb = graphical_model(xb)
    # Pick index with highest probability
    _, preds  = torch.max(yb, dim=1)
    # Retrieve the class label
    return preds
```
